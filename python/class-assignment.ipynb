4. NUMBERS FILTER SYSTEM (PRIME, ODD, EVEN)

```python
while True:
    print("=== Number Filter Program ===")

    start = int(input("Enter the start of the range: "))
    end = int(input("Enter the end of the range: "))

    print("\nChoose an option:")
    print("1. Prime numbers")
    print("2. Odd numbers")
    print("3. Even numbers")
       

    choice = input("Enter your choice (1/2/3): ")

    print("\nNumbers in the range:")
    
    for num in range(start, end + 1):
        if choice == "1":
            if num > 1:
                prime = True
                for i in range(2, num):
                    if num % i == 0:
                        prime = False
                        break
                if prime:
                    print(num)
        elif choice == "2":
            if num % 2 != 0:
                print(num)

        elif choice == "3":
            if num % 2 == 0:
                print(num)
        else:
            print("Invalid choice!")
            break
    print("\nDo you want to run the program again?")
    again = input("Enter Y to continue or any other key to exit: ")
    if again.lower() != "y":
        print("Goodbye!")
        break
```

5. CLASS ,FUNCTION ,OBJECT BASED ON CAR TYPE 

```python
class Car:
    """
        Car Class
        -----------
        Demonstrates:
        ➤ Class
        ➤ Constructor
        ➤ Initialization of attributes
        ➤ Methods
        ➤ Object creation
    """

    def __init__(self, brand, year, color, price, car_type):
        # Constructor: initialize object data
        self.brand = brand
        self.year = int(year)
        self.color = color
        self.price = float(price)
        self.car_type = car_type   # NEW ATTRIBUTE

    # Method to calculate car age
    def age(self):
        current_year = 2026
        return current_year - self.year

    # Method to display details
    def show_details(self):
        print("\n=======================")
        print("      CAR DETAILS      ")
        print("=======================")
        print(f"Brand           : {self.brand}")
        print(f"Production Year : {self.year}")
        print(f"Color           : {self.color}")
        print(f"Type of Car     : {self.car_type}")
        print(f"Price           : ${self.price:,.2f}")
        print(f"Car Age         : {self.age()} years")
        print("=======================\n")


# -------- USER INPUT --------
print("WELCOME TO CAR INFORMATION SYSTEM\n")

print("Examples of car types: Sedan, SUV, Hatchback, Coupe, Electric, Sports, Truck\n")

brand = input("Enter car brand: ")
year = input("Enter production year: ")
color = input("Enter car color: ")
price = input("Enter car price: ")
car_type = input("Enter type of car: ")

# -------- OBJECT CREATION --------
car1 = Car(brand, year, color, price, car_type)

# -------- DISPLAY --------
car1.show_details()
```

6. MULTIPLOT PROGRAM 

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# 2. Calculate y values
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# 3. Clean up Tangent to prevent vertical lines at asymptotes
# We replace values outside our "viewing area" with NaN
y_tan[np.abs(y_tan) > 10] = np.nan 

# 4. Create the plot
plt.figure(figsize=(12, 6))

# Plotting all three lines
plt.plot(x, y_sin, label="sin(x)", color="blue", linewidth=2)
plt.plot(x, y_cos, label="cos(x)", color="green", linewidth=2)
plt.plot(x, y_tan, label="tan(x)", color="red", linestyle="--", alpha=0.7)

# 5. Set the limits (Crucial for visibility)
plt.ylim(-2.5, 2.5) 

# 6. Formatting the graph
plt.axhline(0, color='black', linewidth=1) # X-axis
plt.axvline(0, color='black', linewidth=1) # Y-axis
plt.title("Combined Trigonometric Functions")
plt.xlabel("x (radians)")
plt.ylabel("Value")
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc="upper right")

plt.show()
```

7. STACK OPERATION

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = [None] * size
        self.top = -1

    # PUSH operation
    def push(self, item):
        if self.top == self.size - 1:
            print("Error! Stack is full")
        else:
            self.top += 1
            self.stack[self.top] = item
            print(f"Pushed {item}")

    # POP operation
    def pop(self):
        if self.top == -1:
            print("Error! Stack is empty")
            return None
        else:
            item = self.stack[self.top]
            self.stack[self.top] = None
            self.top -= 1
            print(f"Popped {item}")
            return item

    # Display stack elements
    def display(self):
        if self.top == -1:
            print("Stack is empty")
        else:
            print("Stack elements:", self.stack[:self.top + 1])


# --------- User Interaction ---------
size = int(input("Enter the size of the stack: "))
stack = Stack(size)

while True:
    print("\nChoose an operation:")
    print("1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        value = input("Enter value to push: ")
        stack.push(value)
    elif choice == '2':
        stack.pop()
    elif choice == '3':
        stack.display()
    elif choice == '4':
        print("Exiting...")
        break
    else:
        print("Invalid choice! Please try again.")
```

8. BANK ACCOUNT PROGRAM

```python
class BankAccount:
    """
    BankAccount Class
    Stores and manages information about a bank account
    """

    # ---------- CONSTRUCTOR ----------
    def __init__(self, account_holder, account_number, account_type, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.account_type = account_type
        self.balance = float(balance)

    # ---------- METHOD: DISPLAY DETAILS ----------
    def display_details(self):
        print("\n===============================")
        print("        ACCOUNT DETAILS        ")
        print("===============================")
        print(f"Account Holder : {self.account_holder}")
        print(f"Account Number : {self.account_number}")
        print(f"Account Type   : {self.account_type}")
        print(f"Balance        : ${self.balance:,.2f}")
        print("===============================\n")

    # ---------- METHOD: DEPOSIT ----------
    def deposit(self, amount):
        amount = float(amount)
        self.balance += amount
        print(f"${amount} deposited successfully.")

    # ---------- METHOD: WITHDRAW ----------
    def withdraw(self, amount):
        amount = float(amount)
        if amount > self.balance:
            print("Insufficient balance! Withdrawal failed.")
        else:
            self.balance -= amount
            print(f"${amount} withdrawn successfully.")


# -------- USER INPUT SECTION --------
print("WELCOME TO SIMPLE BANK ACCOUNT SYSTEM\n")

account_holder = input("Enter account holder name: ")
account_number = input("Enter account number: ")
account_type = input("Enter account type (Savings/Current): ")
balance = input("Enter initial balance: ")

# -------- OBJECT CREATION --------
account1 = BankAccount(account_holder, account_number, account_type, balance)

# -------- DISPLAY ACCOUNT --------
account1.display_details()

# -------- PERFORM TRANSACTIONS --------
deposit_amount = input("Enter amount to deposit: ")
account1.deposit(deposit_amount)

withdraw_amount = input("Enter amount to withdraw: ")
account1.withdraw(withdraw_amount)

# -------- SHOW FINAL DETAILS --------
account1.display_details()
```

9. AREA OF CIRCLE AND TRIANGLE 

```python
# Program to calculate area of Circle or Rectangle based on user's choice

print("AREA CALCULATION PROGRAM")
print("-------------------------")
print("1. Area of Circle")
print("2. Area of Rectangle")

# User choice
choice = input("Enter your choice (1 or 2): ")

# If user chooses circle
if choice == "1":
    radius = float(input("Enter radius of circle: "))
    pi = 3.14159
    area = pi * radius * radius
    print("Area of Circle =", area)

# If user chooses rectangle
elif choice == "2":
    length = float(input("Enter length of rectangle: "))
    width = float(input("Enter width of rectangle: "))
    area = length * width
    print("Area of Rectangle =", area)

# Invalid choice
else:
    print("Invalid choice! Please run the program again.")
```
