2. DATAFRAME (2D)

```python
import pandas as pd

data = {
    "Name": [ "Joseph", "Asha", "John", "Fatma", "Kelvin",  "Sarah", "Said", "Neema", "Juma", "Anna"],
    "Age": [ 23, 26, 25, 28, 22,  24, 30, 26, 29, 27],
    "City": [ "Dar", "Mwanza", "Mwanza", "Dodoma", "Tanga",  "Mbeya", "Zanzibar", "Iringa", "Morogoro", "Tabora"],
    "Profession": [ "Engineer", "Doctor", "Artist", "Teacher", "Developer",  "Pilot", "Chef", "Nurse", "Accountant", "Lawyer"]
}
df = pd.DataFrame(data)
df
```

3. SELECTING DATA

```python
df['Name'] #select one column
```

```python
df[['Name', 'City']] #select multiple columns
```

```python
df.loc[6]
```

4. FILTERING DATA

```python
df[df['Age'] > 24]
```

```python
df[(df['Age'] > 24) & (df['City'] == 'Dar')]
```

5. ADDING & MODIFYING COLUMNS

```python
df['salary'] = [500000, 600000, 550000, 500000, 600000, 550000, 500000, 600000, 550000, 545454]
df.head() #first 5 rows
```

```python
df['Age'] = df['Age'] + 1 # will keep updating for every execution
df.tail()
```

6. REMOVING DATA

```python
df.drop('salary', axis=1, inplace=True) #remove column
df.head()
```

```python
df.drop(1, axis=0, inplace=True)
df
```
